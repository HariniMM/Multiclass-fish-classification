# **Project Summary -**

This project focuses on building an effective deep learning-based system for classifying fish images into multiple species categories. The ability to accurately identify fish species from images has important applications in fisheries management, marine biodiversity research, and the seafood industry. Using a combination of data preprocessing, model training, evaluation, and deployment, the project demonstrates a complete pipeline from raw data to an interactive prediction application.

Dataset and Preprocessing

The dataset consists of images categorized into several fish species, organized neatly into separate folders for training, validation, and testing. The images vary in quantity per class, highlighting the presence of class imbalance — some species have many images, while others have relatively few. To prepare the dataset, image files were loaded using TensorFlow’s ImageDataGenerator, which supports on-the-fly data augmentation such as rotation, flipping, and zooming. This augmentation enhances the model’s robustness by exposing it to varied versions of input images, mitigating overfitting. Additionally, all images were rescaled to a [0,1] range and resized uniformly to 224x224 pixels, a standard input size for many pretrained convolutional neural networks (CNNs).

Model Development

The project explored two main modeling strategies. First, a CNN model was built from scratch to establish a baseline. Then, transfer learning was employed using the powerful ResNet50 architecture pretrained on ImageNet. Transfer learning allows leveraging rich feature representations learned from large datasets, accelerating convergence and improving accuracy even with limited data. The ResNet50 base layers were mostly frozen to retain pretrained weights, while the last layers were fine-tuned on the fish dataset.

To address the class imbalance, class weights were calculated inversely proportional to class frequencies and applied during training, ensuring that minority classes contributed more to the loss function and encouraging the model to learn equally from all classes. Hyperparameter tuning was conducted to optimize learning rate, batch size, and the number of trainable layers, improving model generalization.

Training and Evaluation

Models were trained using categorical crossentropy loss and evaluated on the validation set after each epoch. Early stopping and learning rate reduction callbacks were implemented to prevent overfitting and adapt the learning process dynamically. The best model checkpoint was saved automatically based on validation accuracy.

Performance metrics such as accuracy, precision, recall, F1-score, and confusion matrices were computed for each class, providing a detailed understanding of strengths and weaknesses. Visualization of training history confirmed smooth convergence and absence of major overfitting.

Statistical Analysis and Visualization

To understand the dataset better, exploratory data analysis was performed including class distribution plots, correlation heatmaps, and pair plots. Hypothesis testing (ANOVA, t-tests, and correlation tests) was conducted to validate assumptions about the data distribution and relationships, reinforcing confidence in the modeling approach.

Deployment

For practical use, the best performing ResNet50 model was saved in HDF5 format and integrated into a Streamlit web application. This app allows users to upload fish images and instantly receive predicted species labels along with confidence scores. The intuitive interface enhances accessibility for stakeholders who may not have technical expertise, facilitating real-world application of the model.

Conclusion and Future Work

This project demonstrates a full-cycle solution for multiclass fish image classification, leveraging modern deep learning techniques and best practices in data handling. The integration of statistical analysis, class imbalance handling, and model fine-tuning contributed to robust performance. Deployment as a Streamlit app showcases the potential for practical impact.

Future enhancements could include expanding the dataset with more species and images, experimenting with other architectures like EfficientNet or vision transformers, and adding explainability features to interpret model decisions. Incorporating real-time video classification and mobile deployment are other promising directions to broaden application scope.

### 1. Handling Missing Values

```python
# Handling Missing Values & Missing Value Imputation
import os
from PIL import Image

def check_and_remove_corrupt_images(dataset_dir):
    removed_files = []
    for split in ['train', 'val', 'test']:
        split_dir = os.path.join(dataset_dir, split)
        for class_name in os.listdir(split_dir):
            class_dir = os.path.join(split_dir, class_name)
            if not os.path.isdir(class_dir):
                continue
            for img_file in os.listdir(class_dir):
                img_path = os.path.join(class_dir, img_file)
                try:
                    with Image.open(img_path) as img:
                        img.verify()
                except (IOError, SyntaxError) as e:
                    print(f"Removing corrupt image: {img_path}")
                    removed_files.append(img_path)
                    os.remove(img_path)
    print(f"Total corrupt images removed: {len(removed_files)}")


dataset_dir = "/content/drive/MyDrive/data"
check_and_remove_corrupt_images(dataset_dir)
```

#### What all missing value imputation techniques have you used and why did you use those techniques?

In this project, the dataset primarily consists of image files, so missing values were handled in the context of missing or unreadable image files rather than empty cells in tabular data. The first step involved detecting and removing corrupt images using the Python Imaging Library (PIL) to ensure only valid data was fed into the model. Since image data cannot be directly imputed like numerical or categorical values, traditional missing value imputation techniques such as mean, median, or mode filling were not applicable. Instead, to mitigate the impact of removed or missing images—especially in classes with fewer samples—data augmentation techniques such as rotation, zoom, flipping, and shifting were applied to artificially expand the dataset. This approach preserved class diversity, improved model robustness, and addressed potential imbalances caused by missing or discarded images.



### 2. Handling Outliers

```python
# Handling Outliers & Outlier treatments
import os
from PIL import Image

def detect_and_handle_outliers(dataset_dir, min_width=100, min_height=100, max_width=2000, max_height=2000):
    removed_files = []
    for split in ['train', 'val', 'test']:
        split_dir = os.path.join(dataset_dir, split)
        for class_name in os.listdir(split_dir):
            class_dir = os.path.join(split_dir, class_name)
            if not os.path.isdir(class_dir):
                continue
            for img_file in os.listdir(class_dir):
                img_path = os.path.join(class_dir, img_file)
                try:
                    with Image.open(img_path) as img:
                        width, height = img.size

                        if width < min_width or height < min_height or width > max_width or height > max_height:
                            print(f"Removing outlier image due to size: {img_path} ({width}x{height})")
                            removed_files.append(img_path)
                            os.remove(img_path)
                except Exception as e:
                    print(f"Error processing {img_path}: {e}")
                    removed_files.append(img_path)
                    os.remove(img_path)
    print(f"Total outlier images removed: {len(removed_files)}")

# Usage
dataset_dir = "/content/drive/MyDrive/data"
detect_and_handle_outliers(dataset_dir)
```

##### What all outlier treatment techniques have you used and why did you use those techniques?

Removed images that are too small or too large, based on thresholds chosen to maintain consistent input sizes.

This helps improve model stability and training efficiency, since extremely large or tiny images can cause issues with batch processing or model convergence.

dimensionality reduction is generally not needed as a separate preprocessing step because convolutional neural networks (CNNs) inherently handle high-dimensional image data efficiently.

Why dimensionality reduction is usually unnecessary for image classification:
CNNs are designed to work directly with high-dimensional pixel data, extracting hierarchical features through convolutional layers, pooling, and nonlinear activations. These layers progressively reduce spatial dimensions and focus on the most relevant patterns.

Applying traditional dimensionality reduction techniques (like PCA or t-SNE) on raw image pixels before training could discard important spatial information and degrade model performance.

Transfer learning with pretrained models implicitly performs a kind of dimensionality reduction by transforming images into compact feature representations learned from large datasets.

Dimensionality reduction is more relevant in tabular data or when you have engineered features and want to reduce redundancy or noise, which doesn’t apply directly to raw image data.

## ***5. ML Model Implementation***

### ML Model - 1

```python
import numpy as np
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

num_classes = train_generator.num_classes

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

for layer in base_model.layers[:-20]:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
outputs = Dense(num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1)

batch_size = 16

train_generator.batch_size = batch_size
validation_generator.batch_size = batch_size

history = model.fit(
    train_generator,
    epochs=5,
    validation_data=validation_generator,
    class_weight=class_weights_dict,
    callbacks=[early_stop, reduce_lr],
    steps_per_epoch=train_generator.samples // batch_size,
    validation_steps=validation_generator.samples // batch_size,
    verbose=2
)

test_generator.reset()

pred_probs = model.predict(test_generator, verbose=0)
pred_classes = np.argmax(pred_probs, axis=1)

true_classes = test_generator.classes
class_indices = test_generator.class_indices
inv_class_indices = {v: k for k, v in class_indices.items()}

for i in range(5):
    print(f"Predicted: {inv_class_indices[pred_classes[i]]}, Actual: {inv_class_indices[true_classes[i]]}")
```

#### 2. Cross- Validation & Hyperparameter Tuning

```python
import keras_tuner as kt
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import mixed_precision

policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

def build_model(hp):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    for layer in base_model.layers[:-10]:
        layer.trainable = False

    x = base_model.output
    x = GlobalAveragePooling2D()(x)

    units = hp.Int('units', min_value=64, max_value=512, step=64)
    x = Dense(units, activation='relu')(x)
    outputs = Dense(train_generator.num_classes, activation='softmax', dtype='float32')(x)

    model = Model(inputs=base_model.input, outputs=outputs)

    lr = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='keras_tuner_dir',
    project_name='fish_classifier_tuning'
)

tuner.search(
    train_generator,
    epochs=10,
    validation_data=validation_generator,
    class_weight=class_weights_dict,
    callbacks=[early_stop]
)

best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()
```

##### Which hyperparameter optimization technique have you used and why?

I used Random Search hyperparameter optimization through the Keras Tuner library.

Random Search explores a randomly selected subset of the hyperparameter space instead of exhaustively testing every possible combination like Grid Search. This approach is generally more efficient and effective when dealing with a large number of hyperparameters or wide value ranges, as it can find good configurations faster with fewer trials. Additionally, Random Search allows you to specify the number of trials to control the tuning time, making it well-suited for deep learning models trained on large image datasets where each training run is computationally expensive.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): The mean number of images is the same across all fish species (no significant difference).

Alternate Hypothesis (H₁): At least one fish species has a significantly different mean number of images.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
import pandas as pd
from scipy.stats import f_oneway

data = {
    'class_name': [
        'animal fish', 'animal fish bass', 'fish sea_food black_sea_sprat',
        'fish sea_food gilt_head_bream', 'fish sea_food hourse_mackerel',
        'fish sea_food red_mullet', 'fish sea_food red_sea_bream',
        'fish sea_food sea_bass', 'fish sea_food shrimp',
        'fish sea_food striped_red_mullet', 'fish sea_food trout'
    ],
    'image_count': [1096, 30, 569, 566, 573, 579, 571, 538, 576, 547, 580]
}

df_counts = pd.DataFrame(data)

groups = [df_counts[df_counts['class_name'] == cls]['image_count'] for cls in df_counts['class_name']]

f_stat, p_value = f_oneway(*groups)

f_stat, p_value
```

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA.

##### Why did you choose the specific statistical test?

comparing the means across more than two independent groups.

One-Way ANOVA is designed for this scenario and determines if there is at least one group mean significantly different from the others.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

df_counts['category'] = df_counts['class_name'].apply(lambda x: 'animal fish' if 'animal fish' in x else 'seafood fish')

animal_counts = df_counts[df_counts['category'] == 'animal fish']['image_count']
seafood_counts = df_counts[df_counts['category'] == 'seafood fish']['image_count']

t_stat, p_value = ttest_ind(animal_counts, seafood_counts, equal_var=False)

t_stat, p_value
```

There are two independent groups and we want to check if their means differ significantly.
